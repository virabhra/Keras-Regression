```python
X=df.drop('price', axis=1).values
```

```python
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
import tensorflow as tf
```

```python
Scaller=MinMaxScaler()
```

```python
X_train=Scaller.fit_transform(X_train)
```

```python
X_test=Scaller.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model=Sequential()
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train,batch_size=128, epochs=400, verbose=1,validation_data=(X_test, y_test))
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(12,8))
loss.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
pre=model.predict(X_test)
```

```python
pre
```

```python
mean_squared_error(y_test, pre)
```

```python
mean_absolute_error(y_test, pre)
```

```python
explained_variance_score(y_test, pre)
```

```python
plt.figure(figsize=(18,6))
plt.scatter(y_test, pre)
plt.plot(y_test, y_test, 'red')
```
